```python
class Color:
    EMPTY = 0
    BLACK = 1
    WHITE = 2

class Empty:
    color = Color.EMPTY
    
    def __str__(self):
        return ' '
    
    def get_moves(self, board, x, y):
        raise Exception("Error!")
        

"""Base class for all figures"""  
class ChessMan:
    IMG = None
    def __init__(self, color):
        self.color = color
    
    def __str__(self):
        return self.IMG[0 if self.color == Color.WHITE else 1]

    
    
"""All figures"""    
class Pawn(ChessMan):
    IMG = '♟♙'
    def get_moves(self, board, x, y):
        moves = []
        if self.color == Color.BLACK and y < 7\
        and board.get_color(x, y+1) == Color.EMPTY:
            moves.append([x,y+1])
        return moves

class King(ChessMan):
    IMG = '♚♔'
    
class Queen(ChessMan):
    IMG = '♛♕'

class Horse(ChessMan):
    IMG = '♞♘'
    
class Rook(ChessMan):
    IMG = "♜♖"

class Bishop(ChessMan):
    IMG = '♝♗'

    
"""Chess board"""
class Board:
    def __init__(self):
        self.board = [[Empty() for i in range(8)] for j in range(8)]
        self.board[1][2] = Pawn(Color.BLACK)
        
    def __str__(self):
        res = ''
        for y in range(8):
            res += ''.join(map(str, self.board[y])) + '\n'
        return res
    
    def move(self, xy_from, xy_to):
        self.board[xy_to[1]][xy_to[0]] = self.board[xy_from[1]][xy_from[0]]
        self.board[xy_from[1]][xy_from[0]] = Empty()
    
    def get_moves(self, x, y):
        return self.board[y][x].get_moves(self, x, y)
    
    def get_color(self, x, y):
        return self.board[y][x].color

    
b = Board()    
def set_color(color):
    return '\033[%sm' % color


colors = [0, 46]
res = ''
i = 0
for y in range(8):
    for x in range(8):
        res += set_color(colors[i]) + str(b.board[y][x]) + ' '
        i = 1 - i
    i = 1 - i
    res += set_color(0) + '\n'
print(res)
exit()  
    
b = Board()
print(b)
```
